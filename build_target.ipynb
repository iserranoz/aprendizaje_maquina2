```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('data/credit_record.csv')
df.head(20)
```

```python
df.MONTHS_BALANCE.value_counts()
```

```python
df_sorted = df.sort_values(by=['ID', 'MONTHS_BALANCE'])
df_sorted.head(20)
```

```python
df_grouped = df_sorted.groupby('ID').head(6)
df_grouped.head(20)
```

```python
df_grouped['pago_numero'] = df_grouped.groupby('ID').cumcount() + 1
df_pivot = df_grouped.pivot(index='ID', columns='pago_numero', values='STATUS')
df_pivot.columns = ['STATUS_PAGO1', 'STATUS_PAGO2', 'STATUS_PAGO3','STATUS_PAGO4', 'STATUS_PAGO5', 'STATUS_PAGO6']

df_final = df_pivot.reset_index()
df_final = df_final.fillna(-1)
df_final.head(40)
```

```python
status_cols = ['STATUS_PAGO1', 'STATUS_PAGO2', 'STATUS_PAGO3', 'STATUS_PAGO4', 'STATUS_PAGO5', 'STATUS_PAGO6']

condicion_1 = (df_final[status_cols] == 'X').sum(axis=1) + (df_final[status_cols] == -1).sum(axis=1) >= 3

condicion_2 = df_final[status_cols].apply(lambda row: row.isin(['1', '2', '3', '4', '5']).any(), axis=1)

condicion_3 = (df_final[status_cols].isin(['C', '0']).sum(axis=1) >= 3)

df_final['target'] = np.select(
    [condicion_1, condicion_2, condicion_3], 
    [-1, 0, 1], 
    default='UNDEFINED' 
)
```

```python
df_target = df_final.loc[df_final['target'].isin(['1','0']),['ID', 'target']]
df_target.info()
```

```python
df_target.target.value_counts(normalize=True)
```

```python
df2 = pd.read_csv('data/application_record.csv')
df2.head(10)
```

```python
df_m = pd.merge(df_target, df2, on='ID', how='inner')
df_m.info()
```
